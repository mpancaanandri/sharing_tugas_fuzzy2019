Sistem Pendukung Keputusan Kelayakan Pemberian Bidikmisi Dengan Fuzzy Logic

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

# Fuzifikasi

```python
Penghasilan_Orang_Tua = ctrl.Antecedent(np.arange(0, 3000000, 1), 'Penghasilan_Orang_Tua')
Nilai_UN = ctrl.Antecedent(np.arange(0, 10.5, 1), 'Nilai_UN')
Nilai_Prestasi = ctrl.Antecedent(np.arange(0, 40.5, 1), 'Nilai_Prestasi')
Kelayakan = ctrl.Consequent(np.arange(0, 100.5, 1), 'Kelayakan')
```

```python
Penghasilan_Orang_Tua['rendah'] = fuzz.trapmf(Penghasilan_Orang_Tua.universe, [0, 0, 500000, 1000000])
Penghasilan_Orang_Tua['sedang'] = fuzz.trimf(Penghasilan_Orang_Tua.universe, [500000, 1000000, 1500000])
Penghasilan_Orang_Tua['tinggi'] = fuzz.trapmf(Penghasilan_Orang_Tua.universe, [1500000, 2500000, 3000000, 3000000])
```

```python
Penghasilan_Orang_Tua.view()
```

```python
Nilai_UN['rendah'] = fuzz.trapmf(Nilai_UN.universe, [0, 0, 4, 6])
Nilai_UN['sedang'] = fuzz.trimf(Nilai_UN.universe, [5, 6, 8])
Nilai_UN['tinggi'] = fuzz.trapmf(Nilai_UN.universe, [7, 9.5, 10, 10])
```

```python
Nilai_UN.view()
```

```python
Nilai_Prestasi['rendah'] = fuzz.trapmf(Nilai_Prestasi.universe, [0, 0, 10, 15])
Nilai_Prestasi['sedang'] = fuzz.trimf(Nilai_Prestasi.universe, [10, 20, 30])
Nilai_Prestasi['tinggi'] = fuzz.trapmf(Nilai_Prestasi.universe, [20, 30, 40, 40])
```

```python
Nilai_Prestasi.view()
```

```python
Kelayakan['rendah'] = fuzz.trapmf(Kelayakan.universe, [0, 0, 10, 100])
Kelayakan['besar'] = fuzz.trapmf(Kelayakan.universe, [10, 90, 100, 100])
```

```python
Kelayakan.view()
```

# Rule Base

```python
rule1 = ctrl.Rule(Penghasilan_Orang_Tua['rendah'] & Nilai_UN['rendah']& Nilai_Prestasi['rendah'], Kelayakan['rendah'])
rule2 = ctrl.Rule(Penghasilan_Orang_Tua['rendah'] & Nilai_UN['rendah']& Nilai_Prestasi['sedang'], Kelayakan['rendah'])
rule3 = ctrl.Rule(Penghasilan_Orang_Tua['rendah'] & Nilai_UN['rendah']& Nilai_Prestasi['tinggi'], Kelayakan['rendah'])
rule4 = ctrl.Rule(Penghasilan_Orang_Tua['rendah'] & Nilai_UN['sedang']& Nilai_Prestasi['rendah'], Kelayakan['rendah'])
rule5 = ctrl.Rule(Penghasilan_Orang_Tua['rendah'] & Nilai_UN['sedang']& Nilai_Prestasi['sedang'], Kelayakan['besar'])
rule6 = ctrl.Rule(Penghasilan_Orang_Tua['rendah'] & Nilai_UN['sedang']& Nilai_Prestasi['tinggi'], Kelayakan['besar'])
rule7 = ctrl.Rule(Penghasilan_Orang_Tua['rendah'] & Nilai_UN['tinggi']& Nilai_Prestasi['rendah'], Kelayakan['besar'])
rule8 = ctrl.Rule(Penghasilan_Orang_Tua['rendah'] & Nilai_UN['tinggi']& Nilai_Prestasi['sedang'], Kelayakan['besar'])
rule9 = ctrl.Rule(Penghasilan_Orang_Tua['rendah'] & Nilai_UN['tinggi']& Nilai_Prestasi['tinggi'], Kelayakan['besar'])
rule10 = ctrl.Rule(Penghasilan_Orang_Tua['sedang'] & Nilai_UN['rendah']& Nilai_Prestasi['rendah'], Kelayakan['rendah'])
rule11 = ctrl.Rule(Penghasilan_Orang_Tua['sedang'] & Nilai_UN['rendah']& Nilai_Prestasi['sedang'], Kelayakan['rendah'])
rule12 = ctrl.Rule(Penghasilan_Orang_Tua['sedang'] & Nilai_UN['rendah']& Nilai_Prestasi['tinggi'], Kelayakan['besar'])
rule13 = ctrl.Rule(Penghasilan_Orang_Tua['sedang'] & Nilai_UN['sedang']& Nilai_Prestasi['rendah'], Kelayakan['besar'])
rule14 = ctrl.Rule(Penghasilan_Orang_Tua['sedang'] & Nilai_UN['sedang']& Nilai_Prestasi['sedang'], Kelayakan['besar'])
rule15 = ctrl.Rule(Penghasilan_Orang_Tua['sedang'] & Nilai_UN['sedang']& Nilai_Prestasi['tinggi'], Kelayakan['besar'])
rule16 = ctrl.Rule(Penghasilan_Orang_Tua['sedang'] & Nilai_UN['tinggi']& Nilai_Prestasi['rendah'], Kelayakan['besar'])
rule17 = ctrl.Rule(Penghasilan_Orang_Tua['sedang'] & Nilai_UN['tinggi']& Nilai_Prestasi['sedang'], Kelayakan['besar'])
rule18 = ctrl.Rule(Penghasilan_Orang_Tua['sedang'] & Nilai_UN['tinggi']& Nilai_Prestasi['tinggi'], Kelayakan['besar'])
rule19 = ctrl.Rule(Penghasilan_Orang_Tua['tinggi'] & Nilai_UN['rendah']& Nilai_Prestasi['rendah'], Kelayakan['rendah'])
rule20 = ctrl.Rule(Penghasilan_Orang_Tua['tinggi'] & Nilai_UN['rendah']& Nilai_Prestasi['sedang'], Kelayakan['rendah'])
rule21 = ctrl.Rule(Penghasilan_Orang_Tua['tinggi'] & Nilai_UN['rendah']& Nilai_Prestasi['tinggi'], Kelayakan['rendah'])
rule22 = ctrl.Rule(Penghasilan_Orang_Tua['tinggi'] & Nilai_UN['sedang']& Nilai_Prestasi['rendah'], Kelayakan['rendah'])
rule23 = ctrl.Rule(Penghasilan_Orang_Tua['tinggi'] & Nilai_UN['sedang']& Nilai_Prestasi['sedang'], Kelayakan['rendah'])
rule24 = ctrl.Rule(Penghasilan_Orang_Tua['tinggi'] & Nilai_UN['sedang']& Nilai_Prestasi['tinggi'], Kelayakan['rendah'])
rule25 = ctrl.Rule(Penghasilan_Orang_Tua['tinggi'] & Nilai_UN['tinggi']& Nilai_Prestasi['rendah'], Kelayakan['rendah'])
rule26 = ctrl.Rule(Penghasilan_Orang_Tua['tinggi'] & Nilai_UN['tinggi']& Nilai_Prestasi['sedang'], Kelayakan['rendah'])
rule27 = ctrl.Rule(Penghasilan_Orang_Tua['tinggi'] & Nilai_UN['tinggi']& Nilai_Prestasi['tinggi'], Kelayakan['besar'])
```

```python
rule1.view()
```

# Rule Evaluation

```python
Kelayakan_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20, rule21, rule22, rule23, rule24, rule25, rule26, rule27])
```

# Defuzzifikasi

```python
layak = ctrl.ControlSystemSimulation(Kelayakan_ctrl)
```

```python
layak.input['Penghasilan_Orang_Tua'] = 100000
layak.input['Nilai_UN'] = 2
layak.input['Nilai_Prestasi'] = 10
```

```python
layak.compute()
print (layak.output['Kelayakan'])
```

```python
Kelayakan.view(sim=layak)
```
